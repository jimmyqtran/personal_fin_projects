<p>Within the world of investment analysis, four probability distributions tend to be used extensivelyâ€”the uniform, binomial, normal, and lognormal. For the following analysis, we will focus particularly on the latter two, namely the normal and lognormal distributions.</p>
<p>Why do we care about the normal distribution in investment analysis? To put it shortly, we tend to assume stock returns follow a normal distribution with the following parameters: mean and standard deviation that model expected average returns and return volatility respectively. </p>
<p>How about the lognormal distribution? We assume stock prices to follow a lognormal distribution. We know that the lognormal distribution has a lower bound of zero as well as skew to the right side. As we recall, stock prices cannot take on a nonnegative value as stocks themselves represent a shares of ownership in a company. Hence, the lognormal distribution sounds like a good distribution to model stock prices on. For a slightly more detailed discussion  of these two distributions with regards to stock prices and returns, go <a href="url">here</a>.</p>

```r
# Load relevant packages
library(tidyverse, warn.conflicts = FALSE)
library(ggthemes)
library(scales)

# Load the dataset from Yahoo Finance
SPY <- read.csv("datasets/SPY_prices.csv")
SPY_div <- read.csv("datasets/SPY_dividends.csv")

# In lieu of Vlookup in Excel, use merge() to combine dividend data with corresponding dates in SPY.df
SPY.df <- merge(x = SPY, y = SPY_div, all.x = TRUE)

# Remove NA from your dividend calculations to avoid calculation headache later
SPY.df$Dividends[is.na(SPY.df$Dividends)] <- 0

# Calculate your returns
## diff() gives lagged difference
## close([-length(Close)]) gives the last observation
SPY.df <- SPY.df %>% mutate(returns = (diff(Close)+Dividends)/Close[-length(Close)])
```

```r
# Graph price on date
prices <- ggplot(data = SPY.df, aes(x = as.Date(Date), y = Close)) + geom_line(colour="blue") + 
  scale_x_date(labels = date_format("%Y"), breaks = date_breaks("5 years"), limits = c(as.Date("1990-1-1"), as.Date("2020-1-1"))) + 
  theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1))

# View the price plot
prices + labs(title ="SPY Price Time Series", x="Time (Years)", y = "Closing Price (USD)") + theme_bw() +
  scale_y_continuous(breaks = c(seq(0,300, by = 50)), limits = c(0,300))

# Graph returns on date
returns <- ggplot(data = SPY.df, aes(x = as.Date(Date), y = returns*100)) + geom_line(colour="blue") + 
  scale_x_date(labels = date_format("%Y"), breaks = date_breaks("2 years"), limits = c(as.Date("1990-1-1"), as.Date("2020-1-1"))) + 
  theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1))

# View the returns plot
returns + labs(title ="SPY Return Time Series", x="Time (Years)", y = "Returns (%)") + theme_bw()
```

```r
# Create your CDF, 1-CDF plot using ordered returns
SPY.df <- SPY.df %>% mutate(count = seq(1:nrow(SPY.df))) %>% 
  mutate(CDF = count/(nrow(SPY.df)+1)) %>% 
  mutate(returns_order = sort(returns)) %>% 
  mutate(one_CDF = ifelse(returns_order >= 0, 1 - CDF, CDF))

# Create theoretical normal CDF using returns, average of returns, and standard deviation of returns
SPY.df <- SPY.df %>% mutate(normal = pnorm(returns_order, mean = mean(SPY.df$returns), 
                                           sd = sd(SPY.df$returns), log = FALSE)) %>% 
  mutate(one_normal = ifelse(returns_order >= 0, 1 - normal, normal))

# Clone data frame to use for last part of the question
SPY.df2 <- SPY.df
SPY.df <- SPY.df %>% gather(`one_CDF`, `one_normal`, key = "plots", value = "return_values")

# Plot the theoretical normal CDF and CDF, 1-CDF on returns
graph2 <- ggplot(data = SPY.df, aes(x=returns_order, y = return_values, group = plots)) + geom_point(aes(colour = plots)) + 
    scale_y_log10(limits = c(0.0001,1)) + 
    theme_bw() +
    scale_colour_ptol()

# Plot the graph
graph2
```

```r
# Isolate the proportion of returns relevant to "fat tails"
SPY.df2 <- SPY.df2 %>% mutate(dependent = (returns_order-mean(returns))/sd(returns)) %>%
  mutate(ln_return = ifelse(abs(dependent)>2,log(abs(dependent)),NA))

# Plot the ln(1,1-CDF) on ln(abs(returns)) relevant to "fat tails"
graph3 <- SPY.df2 %>% ggplot(aes(y=log(one_CDF), x = ln_return)) + geom_point() + 
  geom_smooth(method = "lm", se=FALSE) + 
  scale_y_continuous(limits=c(-10,-2)) + 
  scale_x_continuous(limits=c(0,3)) +
  labs(title = "SPY Cumulative Return Distribution: Power Law Fat Tail", 
       y = "ln(CDF,1-CDF)", x = "ln(abs(standardized returns))")

# View the plot
graph3 + theme_bw()
```

```r
# Find the R-squared value and coefficient of interest pertaining to the Gabaix reading on Power Law
gabaix <- lm(SPY.df2, formula = log(one_CDF) ~ ln_return)
summary(gabaix)
```
